```python
import numpy as np
import matplotlib.pyplot as plt

#Korrekturfaktor 68%
t68 = 1;
#Korrekturfaktor 95%
t95 = 1.96;

#Daten kurze Seite einlesen
data2 = np.genfromtxt('messungA3/dinA4-kurz.csv', delimiter=";", skip_header=1000, skip_footer=500,
                         usecols=(1), converters={1: lambda x: float(x.decode('utf-8').replace(',', '.'))})

#Daten lange Seite einlesen
data1 = np.genfromtxt('messungA3/dinA4-lang.csv', delimiter=";", skip_header=1000, skip_footer=500,
                         usecols=(1), converters={1: lambda x: float(x.decode('utf-8').replace(',', '.'))})

print("DinA4 lang: ")    
#Mittelwert Lange Seite
data1mean = np.mean(data1)
#Standardabweichung lange Seite
data1std = np.std(data1)
#Korrektur durch anzahl messungen
data1stdcorrected = np.std(data1) / np.sqrt(data1.size)
#print("Datapoints " + str(data1.size))
#print("Mean " + str(data1mean))
#print("Std " + str(data1std))
#print("std corrected " + str(data1stdcorrected))
print("Durch > 90000 Datenpunkte Bei P = 68,26% t = 1 und bei P = 95% t = 1.96")

print("also ergibt sich ")
print()
print("68,26%: " + str(data1mean) + " +- "+ str(t68) + " * " + str(data1stdcorrected) + " * 1 \n=> " + str(data1mean) + "V +- " + str(t68*data1stdcorrected) + "V")
print("95%: " + str(data1mean) + " +- " + str(t95) + " * " + str(data1stdcorrected) + " * 2 \n=> " + str(data1mean) + "V +- " + str(t95*data1stdcorrected*2) + "V")
print()


print("dasselbe fuer die andere seite: ")
print("DinA4 kurz: ")
#Mittelwert kurze Seite
data2mean = np.mean(data2)
#Standardabweichung kurze Seite
data2std = np.std(data2)
#Korrektur durch anzahl messungen
data2stdcorrected = np.std(data2) / np.sqrt(data2.size)
#print("Datapoints " + str(data2.size))
#print("Mean " + str(data2mean))
#print("Std " + str(data2std))
#print("std corrected " + str(data2stdcorrected))
print("Durch > 90000 Datenpunkte Bei P = 68,26% t = 1 und bei P = 95% t = 1.96")
print("daraus folgt ")
print()
print("68,26%: " + str(data2mean) + " +- "+ str(t68) + " * " + str(data2stdcorrected) + " * 1 \n=> " + str(data2mean) + "V +- " + str(t68*data2stdcorrected*1) + "V")
print("95%: " + str(data2mean) + " +- " + str(t95) + " * " + str(data2stdcorrected) + " * 2 \n=> " + str(data2mean) + "V +- " + str(t95*data2stdcorrected*2) + "V")
print()


#Berechnung von a zur langen Seite
a1 = np.log(29.7) / np.log(data1mean)
#Berechnung von b zur langen Seite
b1 = np.log(29.7) - (a1 * np.log(data1mean))
#Berechnung des Messfehlers.
v1 = (np.exp(b1) * a1 * (data1mean**(a1 - 1)))
#Umkehrung der Logarithmierung
res1 = np.exp(b1) * data1mean**(a1)

print("Langes DinA4 mit 68,26%: " + str(round(res1, 2)) + "cm +- " + str((v1*data1stdcorrected) * t68 * -1) + "cm")
print("Langes DinA4 mit 95%: " + str(round(res1, 2)) + "cm +- " + str((v1*data1stdcorrected) * t95 * -1) + "cm")


#Berechnung von a zur kurzen Seite
a2 = np.log(21) / np.log(data2mean)
#Berechnung von b zur langen Seite
b2 = np.log(21) - (a2 * np.log(data2mean))
#Berechnung des Messfehlers.
v2 = (np.exp(b2) * a2 * (data2mean**(a2 - 1)))
#Umkehrung der Logarithmierung
res2 = np.exp(b2) * data2mean**(a2)

print("Kurzes DinA4 mit 68,26%: " + str(round(res2, 2)) + "cm +- " + str((v2*data2stdcorrected) * t68 * -1) + "cm")
print("Kurzes DinA4 mit 95%: " + str(round(res2, 2)) + "cm +- " + str((v2*data2stdcorrected) * t95 * -1) + "cm")


#2sds
```
